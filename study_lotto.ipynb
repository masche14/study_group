# 인원 및 파트 설정

```python
# 스터디 그룹 멤버
org_study_members=['안응민', '정선진', '박은호', '신예훈', '신동준', '김성민', '신예준']

# 0주차 시험공부
exam_subjects=['자바','자료구조','정보통신','운영체제','C언어','직업과경력','HTML','데이터베이스','리눅스'] 
# 1주차 데이터 모델링의 이해
first_week=['데이터 모델링의 이해','엔터티','속성','관계 및 식별자']
# 2주차 데이터 모델과 성능
second_week=['성능 데이터 모델링의 개요','정규화/반정규화 성능','대량 데이터에 따른 성능','데이터 베이스 구조와 성능','데이터 베이스 구조와 성능','분산 데이터베이스와 성능']
# 3주차 SQL 기본 및 활용
third_week=['관계형 데이터베이스의 개요','DDL/DML/TCL/DCL','조인(JOIN)','그룹함수','윈도우함수']
# 4주차 SQL 최적화 기본 원리
fourth_week=['옵티마이저와 실행계획','인덱스 기본','조인 수행 원리']

sqld=[]
sqld.append(first_week)
sqld.append(second_week)
sqld.append(third_week)
sqld.append(fourth_week)
```

# 발표 파트 랜덤 지정

```python
study_members=org_study_members[1:] # 안응민 제외한 인원만 sqld 발표

week=int(input("주차를 입력하세요. : ")) # 주차 입력

test_list=sqld[week-1] 

print(f'목차 개수 : {len(test_list)}')

# random 패키지를 활용한 발표를 맡을 스터디 그룹 멤버 및 발표 파트 랜덤추출
if len(study_members)>len(test_list): # 파트 수가 인원수 보다 적을 경우 인원수 - 파트 수 차이만큼 발표에서 구제
    diff_cnt=len(study_members)-len(test_list)
    print(f'{diff_cnt}명 구제')
    random_number=len(test_list)
    member=random.sample(study_members, random_number)
    except_members=', '.join([i for i in study_members if i not in member])
    print(f'구제 명단 : {except_members}')
    part=random.sample(test_list, random_number)
else:
    random_number=len(study_members)
    member=random.sample(study_members, random_number)
    part=random.sample(test_list, random_number)

quiz_dict={'member':member
           , 'part':part
          }

quiz_df=pd.DataFrame(quiz_dict)
# dataframe_image.export(quiz_df, f'{week}주차 배정.png') # 데이터프레임을 이미지로 저장하여 해당 이미지는 카카오톡으로 공유할 예정
quiz_df
```
